```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
dataset=pd.read_csv("/content/Churn_Modelling.csv")
```

```python
# convert categerical feattures to numerical featurs
x_train=dataset.iloc[:, :-1]
y_train=dataset.iloc[:,-1]
```

```python
x_train.shape
```

```python
x_train.describe()
```

```python


cat_feat=["Surname","Geography","Gender"]
# label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat_feat:
  x_train[i]=le.fit_transform(x_train[i])
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
```

```python
# split data train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_train,y_train,test_size=0.3,random_state=1)
```

```python
# use of ANN
import tensorflow as tf
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann=Sequential()
from tensorflow.keras.regularizers import L2  # âœ… Correct import (lowercase 'l2')
```

```python
ann.add(Dense(units=6, input_dim=13, activation="relu", kernel_regularizer=L2(0.01)))
ann.add(Dense(units=6,activation="relu",kernel_regularizer=L2(0.01)     )   )
ann.add(Dense(units=1,activation="sigmoid"))
```

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=15,epochs=20,validation_data=(x_test,y_test))
```

```python
# calculate total accuraacy
from sklearn.metrics import accuracy_score
y_pred=ann.predict(x_test)
```

```python


test=[]
for i in y_pred:
  if i >  0.5:
    test.append(1)
  else:
    test.append(0)
y_pred=np.array(test)
accuracy_score(y_test,y_pred)
```
